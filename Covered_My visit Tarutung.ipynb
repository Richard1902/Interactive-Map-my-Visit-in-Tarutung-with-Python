# Coloring markers

Feature engineering: creating a feature for coloring the markers

```python
def select_marker_color(row):
    if row['lived'] == 'yes':
        return 'pink'
    elif row['lived'] == 'no' and row['visited'] == 'yes':
        return 'purple'
    return 'blue'
```

```python
cities['colors'] = cities.apply(select_marker_color, axis=1)
cities.head(3)
```

Visualizing colored marker

```python
world_tarutungarea_colored = folium.Map(
    zoom_start=10,
    location=[2.0845867538935696, 99.04713597668687]
)

for _, city in cities.iterrows():
    folium.Marker(
        location=[city['latitude'], city['longitude']],
        popup=city['name'],
        tooltip=city['name'],
        icon=folium.Icon(color=city['colors'], prefix='fa', icon='circle')
    ).add_to(world_tarutungarea_colored)
    
world_tarutungarea_colored
```

```python
world_tarutungarea_colored.save('maps_htmls/world_tarutungarea_colored.html')
```
